```python
# loc - ...
# iloc - satır ve sütun sayısı

X= traindata.iloc[:, [6]].values
X
```

```python
# k means - model kuruldu 300 deneme
# her adımda 10 farklı küme yoğunluğu analiz edildi

# düze indiği alan en iyi sonuç(x ekseni 5: 5 küme uygundur)

from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i,init='k-means++',max_iter=300,n_init=10,random_state=0)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow Curve')
plt.show()
```

```python
y= testdata.iloc[:, [6]].values
```

```python
# 5 kümeye ayrılacak, 300

from sklearn.cluster import KMeans

kmeans=KMeans(n_clusters=5,init='k-means++',max_iter=300,n_init=10,random_state=0)
y_kmeans=kmeans.fit_predict(y)
y_kmeans
```

```python
# başarıyı hesaplama

from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y_true=y, y_pred=y_kmeans)
print("Doğruluk: % {:10.2f}".format(accuracy*100))
```
